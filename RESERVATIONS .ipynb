```python
def query(sql):
    """Execute a SQL query and return the result as a pandas DataFrame."""
    return pd.read_sql_query(sql, conn)

print("=== PASSENGERS ===")
display(query("SELECT * FROM passengers LIMIT 5"))
print("=== FLIGHTS ===")
display(query("SELECT * FROM flights LIMIT 5"))
print("=== BOOKINGS ===")
display(query("SELECT * FROM bookings LIMIT 5"))
```

```python
result = query("""
 Select flight_number,destination,departure_time,  price_eur from Flights 
Where origin = 'Nice'
Order by departure_time
""")

display(result)

```

```python
result = query("""
Select flight_number, destination, price_eur  from flights where price_eur<100
Order By price_eur ASC
""")
display(result)

```

```python
result = query("""

Select f.destination, 
    SUM(f.price_eur) as total_revenue, 
    COUNT(b.id) as nb_bookings
FROM bookings b 
INNER JOIN flights f ON b.flight_id = f.id 
WHERE b.status = 'confirmed'
GROUP BY f.destination
ORDER BY total_revenue DESC
""")
display(result)

```

```python
result = query("""
SELECT f.destination, COUNT(b.id) as nb_bookings
FROM bookings b 
INNER JOIN flights f ON b.flight_id = f.id
Where b.status ='confirmed'
Group By f.destination
HAVING COUNT(b.id)> 1
ORDER BY nb_bookings DESC
""")
display(result)

```

```python
result = query("""
SELECT p.first_name, p.last_name, b.seat_class, b.seat_number
FROM bookings b 
INNER JOIN passengers p ON b.passenger_id = p.id
INNER JOIN flights f ON b.flight_id = f.id
WHERE f.flight_number = 'AF5501'
AND b.status = 'confirmed'
""")
display(result)

```

```python
result = query("""
SELECT first_name, p.last_name 
FROM passengers p
LEFT JOIN bookings b ON p.id = b.passenger_id
WHERE b.id IS NULL
""")
display(result)

```

```python
result = query("""
SELECT flight_number, destination, price_eur
FROM flights
WHERE price_eur> (SELECT AVG(price_eur) FROM flights)
ORDER BY price_eur DESC
""")
display(result)


```

```python
result = query("""
SELECT p.first_name, p.last_name, COUNT(b.id) as nb_bookings 
FROM passengers p
INNER JOIN bookings b ON p.id = b.passenger_id 
GROUP BY p.id, p.first_name, p.last_name
HAVING COUNT(b.id) >= 2
ORDER BY nb_bookings DESC

""")
display(result)

```
